```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
expenditures = pd.read_csv('Expenditures_2018_to_Current_Year.csv')
vendors = pd.read_csv('Vendor_List.csv')
```

```python
#Drop invoice date this column is completely nan
expenditures.drop('invoice_date',axis =1, inplace = True)
```

```python
#expenditures['Check_Date']
expenditures['Check_Date'] = pd.to_datetime(expenditures['Check_Date'])
expenditures['Check_Date'] = expenditures['Check_Date'].dt.date
expenditures['Check_Date'] = pd.to_datetime(expenditures['Check_Date'])
```

```python
vendors.columns = vendors.columns.str.lower().str.strip()
```

```python
vendors.columns.str.strip()
```

```python
expenditures.columns = expenditures.columns.str.lower().str.strip()
expenditures.columns
```

```python
engine = create_engine("postgresql://postgres:admin@localhost:5432/Naperville Finances")
```

```python
vendors.to_sql('vendors', engine, index = False)
```

```python
expenditures.to_sql('expenditures', engine, index = False)
```
